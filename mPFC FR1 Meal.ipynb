```python
import meals as ml
import tools as tl
import intervals as invs
import paths
from preprocessing import read_csv_clean

files = paths.fr1_fent
```

```python
fent_pellet_avg = []
fent_meal_count = []
```

```python
files
```

```python
for file in files:
    print(file)
    # data = read_csv_clean(file)
    data = read_csv_clean(file)
    meal = ml.find_meals(data)
    group = ml.find_pellet_frequency(data)
    num = tl.get_bhv_num(file)[0]
    ml.graphing_cum_count(data, meal, None, num)
    ml.graph_pellet_frequency(group, None, num)
    temp = ml.average_pellet(group)
    fent_pellet_avg.append(temp)
    avg = round(len(meal)/ml.experiment_duration(data), 2)
    fent_meal_count.append(avg)
```

```python
files = paths.fr1_veh
```

```python
veh_pellet_avg = []
veh_meal_count = []
```

```python
for file in files:
    print(file)
    data = read_csv_clean(file)
    meal = ml.find_meals(data)
    group = ml.find_pellet_frequency(data)
    num = tl.get_bhv_num(file)[0]
    ml.graphing_cum_count(data, meal, None, num)
    ml.graph_pellet_frequency(group, None, num)
    temp = ml.average_pellet(group)
    veh_pellet_avg.append(temp)
    avg = round(len(meal)/ml.experiment_duration(data), 2)
    veh_meal_count.append(avg)
```

### Tests

```python
import intervals as invs

invs.perform_T_test(ctrl=fent_meal_count, exp=veh_meal_count, test_side='greater')
```

```python
invs.perform_T_test(ctrl=fent_pellet_avg, exp=veh_pellet_avg, test_side='greater')
```
