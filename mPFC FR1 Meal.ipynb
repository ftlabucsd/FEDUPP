```python
import meals as ml
import tools as tl
import intervals as invs
import paths
from preprocessing import read_csv_clean, prep_pellet_count

files = paths.fr1_veh
```

```python
fent_pellet_avg = []
fent_meal_count = []
```

```python
for file in files:
    print(file)
    # data = read_csv_clean(file)
    data = read_csv_clean(file, processed=True)
    meal = ml.find_meals(data)
    group = ml.find_pellet_frequency(data)
    ml.graphing_cum_count(data, meal, None, None)
    ml.graph_pellet_frequency(group, None, None)
    temp = ml.average_pellet(group)
    fent_pellet_avg.append(temp)
    avg = round(len(meal)/ml.experiment_duration(data), 2)
    fent_meal_count.append(avg)
```
