```python
import pandas as pd
import paths
```

```python
paths.contigency_flip_ctrl
```

```python
path = '../behavior data integrated/CASK/CASK_data_info.xlsx'
table = pd.DataFrame()
```

```python
mouse_name = []
session = []
file_path = []
types = []

for file in paths.contigency_flip_ctrl:
    parts = file.split('/')
    cutoff = parts[-1].index('.')
    mouse_name.append(parts[-1][:cutoff])
    types.append('Control')
    session.append(parts[-3])
    file_path.append(file)

for file in paths.contigency_flip_cask:
    parts = file.split('/')
    cutoff = parts[-1].index('.')
    mouse_name.append(parts[-1][:cutoff])
    types.append('CASK')
    session.append(parts[-3])
    file_path.append(file)
```

```python
table['Mouse'] = mouse_name
table['FR1/Reversal'] = session
table['File Path'] = file_path
table['Type'] = types
```

```python
table
```

```python
table.to_excel(path, index=False)
```
