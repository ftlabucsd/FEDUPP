```python
import meals as ml
import tools as tl
import intervals as invs

import paths
from preprocessing import read_csv_clean

ctrl_files = paths.fr1_ivsa_ctrl
exp_files = paths.fr1_ivsa_exp
```

```python
ctrl_pellet_avg = []
ctrl_meal_count = []

for file in ctrl_files:
    print(file)
    data = read_csv_clean(file)
    meal = ml.find_meals(data)
    group = ml.find_pellet_frequency(data)
    info = tl.get_bhv_num(file)
    if len(info) == 1:
        ml.graphing_cum_count(data, meal, None, info[0])
        ml.graph_pellet_frequency(group, None, info[0])
    else:
        ml.graphing_cum_count(data, meal, info[0], info[1])
        ml.graph_pellet_frequency(group, info[0], info[1])
    temp = ml.average_pellet(group)
    ctrl_pellet_avg.append(temp)
    avg = round(len(meal)/ml.experiment_duration(data), 2)
    ctrl_meal_count.append(avg)
```

```python
exp_pellet_avg = []
exp_meal_count = []

for file in exp_files:
    print(file)
    data = read_csv_clean(file)
    meal = ml.find_meals(data)
    group = ml.find_pellet_frequency(data)
    info = tl.get_bhv_num(file)
    if len(info) == 1:
        ml.graphing_cum_count(data, meal, None, info[0])
        ml.graph_pellet_frequency(group, None, info[0])
    else:
        ml.graphing_cum_count(data, meal, info[0], info[1])
        ml.graph_pellet_frequency(group, info[0], info[1])
    temp = ml.average_pellet(group)
    exp_pellet_avg.append(temp)
    avg = round(len(meal)/ml.experiment_duration(data), 2)
    exp_meal_count.append(avg)
```
