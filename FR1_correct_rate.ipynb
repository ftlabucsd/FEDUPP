## FR1 Correct Rate

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import meals as ml
```

#### Define preprocess function for read excel

```python
def read_excel_by_sheet(sheet, parent='../behavior data integrated/Adjusted FED3 Data.xlsx'):
    df = pd.read_excel(parent, sheet_name=sheet)

    df = df[['MM:DD:YYYY hh:mm:ss', 'Event', 'Active_Poke',
         'Cum_Sum', 'Percent_Correct']].rename(columns={'MM:DD:YYYY hh:mm:ss': 'Time'}).fillna(0)
    
    df['Percent_Correct'] *= 100
    df['Time'] = pd.to_datetime(df['Time'])
    mask = (df['Percent_Correct'] != 0) & (df['Cum_Sum'] != 0)
    df = df[mask]
    
    return df
```

## Control Group

###  Bhv 2

```python
B2M1 = read_excel_by_sheet(sheet='B2.M2')
B2M2 = read_excel_by_sheet(sheet='B2.M3')
B2M3 = read_excel_by_sheet(sheet='B2.M4')
```

### Learning Rate for Bhv 2

```python
plt.figure(figsize=(15, 6), dpi=90)
sns.lineplot(data=B2M1, x='Time', y='Percent_Correct', label='M1')
sns.lineplot(data=B2M2, x='Time', y='Percent_Correct', label='M2')
sns.lineplot(data=B2M3, x='Time', y='Percent_Correct', label='M3')

plt.grid()
plt.title('Changes in Correction Rate for Control Group 2', fontsize=24)
plt.xlabel('Time', fontsize=16)
plt.ylabel('Correct Rate', fontsize=16)
plt.legend()
legend = plt.legend(title='Mice', fontsize=10)
legend.get_title().set_fontsize(12)
plt.show()
```

```python
plt.figure(figsize=(15, 6), dpi=90)

sns.lineplot(data=B2M1, x='Time', y='Cum_Sum', label='M1')
sns.lineplot(data=B2M2, x='Time', y='Cum_Sum', label='M2')
sns.lineplot(data=B2M3, x='Time', y='Cum_Sum', label='M3')

plt.grid()
plt.title('Cumulative Sum of Pellet for Control Group 2', fontsize=22)
plt.xlabel('Time', fontsize=16)
plt.ylabel('Cumulative Percentage', fontsize=16)
plt.legend()
legend = plt.legend(title='Mice', fontsize=10)
legend.get_title().set_fontsize(12)
plt.show()
```
