```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import meals as ml
import intervals as invs
import numpy as np
```

#### Define preprocess function for read excel

```python
def read_excel_by_sheet(sheet, parent='../behavior data integrated/Adjusted FED3 Data.xlsx'):
    df = pd.read_excel(parent, sheet_name=sheet)

    df = df[['MM:DD:YYYY hh:mm:ss', 'Event', 'Active_Poke',
         'Cum_Sum', 'Percent_Correct']].rename(columns={'MM:DD:YYYY hh:mm:ss': 'Time'}).fillna(0)
    
    df['Percent_Correct'] *= 100
    df['Time'] = pd.to_datetime(df['Time'])
    mask = (df['Percent_Correct'] != 0) & (df['Cum_Sum'] != 0)
    df = df[mask]
    
    return df
```

```python
ctrl_ending_corr = []
cask_ending_corr = []
```

```python
B2M1 = read_excel_by_sheet(sheet='B2.M2')
B2M2 = read_excel_by_sheet(sheet='B2.M3')
B2M3 = read_excel_by_sheet(sheet='B2.M4')
```

```python
ctrl_ending_corr.append(B2M1['Percent_Correct'].tail(1).values[0])
ctrl_ending_corr.append(B2M2['Percent_Correct'].tail(1).values[0])
ctrl_ending_corr.append(B2M3['Percent_Correct'].tail(1).values[0])
ctrl_ending_corr
```

```python
ctrl_mean = np.mean(ctrl_ending_corr)
cask_mean = np.mean(cask_ending_corr)
ctrl_err = np.std(ctrl_ending_corr) / np.sqrt(len(ctrl_ending_corr))
cask_err = np.std(cask_ending_corr) / np.sqrt(len(cask_ending_corr))
groups = ['Control', 'CASK']
```

```python
x = np.arange(len(groups))
width = 0.4
```

```python
# Create the bar plot for averages
plt.figure(figsize=(7, 7))

plt.bar([1, 2], [ctrl_mean, cask_mean], yerr=[ctrl_err, cask_err], capsize=12, tick_label=groups, 
        width=width, color=['lightblue', 'yellow'], alpha=0.8, zorder=1, label=['Control', 'CASK'])

x1 = [1] * len(ctrl_ending_corr)
x2 = [2] * len(cask_ending_corr)
plt.scatter(x1, ctrl_ending_corr, marker='o', color='blue', zorder=2) 
plt.scatter(x2, cask_ending_corr, marker='x', color='orange', zorder=2)

plt.xlabel('Groups', fontsize=14)
plt.ylabel('Correct Rate (%)', fontsize=14)
plt.title('Average Correct Rate of Control and CASK Groups in FR1', fontsize=16)

plt.legend()
plt.show()
```
