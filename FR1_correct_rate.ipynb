### Analysis

```python
invs.perform_T_test(ctrl_ending_corr, cask_ending_corr)
```

```python
invs.MannWhitneyUTest(ctrl_ending_corr, cask_ending_corr)
```

```python
ctrl_mean = np.mean(ctrl_ending_corr)
cask_mean = np.mean(cask_ending_corr)
ctrl_err = np.std(ctrl_ending_corr) / np.sqrt(len(ctrl_ending_corr))
cask_err = np.std(cask_ending_corr) / np.sqrt(len(cask_ending_corr))
groups = ['Control', 'CASK']
```

```python
x = np.arange(len(groups))
width = 0.4
```

```python
# Create the bar plot for averages
plt.figure(figsize=(7, 7))

plt.bar([1, 2], [ctrl_mean, cask_mean], yerr=[ctrl_err, cask_err], capsize=12, tick_label=groups, 
        width=width, color=['lightblue', 'yellow'], alpha=0.8, zorder=1, label=['Control', 'CASK'])

x1 = [1] * len(ctrl_ending_corr)
x2 = [2] * len(cask_ending_corr)
plt.scatter(x1, ctrl_ending_corr, marker='o', color='blue', zorder=2) 
plt.scatter(x2, cask_ending_corr, marker='x', color='orange', zorder=2)

plt.xlabel('Groups', fontsize=14)
plt.ylabel('Correct Rate (%)', fontsize=14)
plt.title('Average Correct Rate of Control and CASK Groups in FR1', fontsize=16)

plt.legend()
plt.show()
```
